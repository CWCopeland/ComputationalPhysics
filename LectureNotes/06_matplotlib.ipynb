### A first plot

For convenience, `pyplot` is customarily imported as `plt`.

```python
import matplotlib.pyplot as plt
import numpy as np
```

To create a line graph, we first define a `Figure` (which can contain one or more sets of `Axes`).

```python
plt.figure(figsize=(10,6))   #figsize argument used to set the plot size
```

Then we can plot a collection of data points by passing them to the `plot()` function.

```python
plt.plot([7,3,9,1])
#plt.show() #may or may not be necesarry
```

Since we didn't provide values for the $x$-axis, Mayplotlib assumes they are sequential, starting with zero.  To avoid this, provide a second set of values to the `plot()` function.

```python
plt.figure()
plt.plot([1,2,4,8],[7,3,9,1])
#plot.show()
```

A title can be added using the `title()` function.

```python
plt.figure()
plt.plot([1,2,4,8],[7,3,9,1])
plt.title("John's Plot", fontsize=20)
#plot.show()
```

A grid can be added (to guide the eye) using the `grid()` function.

```python
plt.figure()
plt.plot([1,2,4,8],[7,3,9,1])
plt.grid(True)
#plot.show()
```

The size of the tick labels can be adjusted using the `tick_params()` function.

```python
plt.figure()
plt.plot([1,2,4,8],[7,3,9,1])
plt.tick_params(axis='both', which='major', labelsize=20)
#plot.show()
```

The $x$[$y$]-axis tick locations and labels can be set using the `xticks()` [`yticks()`] function.  Pass these functions a list of tick positions to change the default behavior.

```python
plt.figure()
plt.xticks(np.linspace(1,8,15))
plt.plot([1,2,4,8],[7,3,9,1])
#plot.show()
```

You can also pass these functions a collection of tick positions and custom labels.

```python
plt.figure()
plt.xticks(range(1,5), ['Jan','Feb','Mar','Apr'])
plt.plot(range(1,5),[7,3,9,1])
#plt.show()
```

You should always provide meaningful axis labels (including units), using the `xlabel()` and `ylabel()` functions.

```python
plt.figure()
plt.xticks(range(1,5), ['Jan','Feb','Mar','Apr'])
plt.xlabel("Month")
plt.ylabel("Temperature [K]")
plt.plot(range(1,5),[7,3,9,1])
#plt.show()
```

The `plot()` function takes a variety of arguments to modify the graph appearance.  Let's draw a wide/dashed/red line with markers at each point.

```python
plt.figure()
plt.plot(range(1,5),[7,3,9,1], color='red', marker='o', linestyle='dashed', linewidth=2, markersize=6)   #marker='o' -> filled circles
#plt.show()
```

The functions `xscale()` and `yscale()` can be used to switch between linear and logarithmic axes.

```python
plt.figure()
plt.plot(range(1,5),[7,3,9,1], color='red', marker='o', linestyle='dashed', linewidth=2, markersize=6)

plt.yscale('log')

#plt.show()
```

To plot to two distributions on a single set of axes, call `plot()` twice.

```python
plt.figure()
plt.plot(range(1,5),[7,3,9,1], color='red', marker='o', linestyle='dashed', linewidth=2, markersize=6)   #marker='o' -> filled circles
plt.plot(range(1,5),[3,4,6,9], color='y', marker='*', linestyle='dashed', linewidth=0.5, markersize=12)
#plt.show()
```

The `legend()` function can be used to create a legend.  To use this function, you must also specify a `label` when plotting each distribution.

```python
plt.figure()
plt.plot(range(1,5),[7,3,9,1], label=r"$\alpha$", color='red', marker='o', linestyle='dashed', linewidth=2, markersize=6)   #use a raw string to define label, to avoid \a being treated as a special character
plt.plot(range(1,5),[3,4,6,9], label=r"$\beta$", color='y', marker='*', linestyle='dashed', linewidth=0.5, markersize=12)
plt.legend(loc='upper left', fontsize='x-large')
#plt.show()
```

The `subplot()` function can be used to put multiple sets of `Axes` in a single `Figure`.  The first argument is the number of subplot rows, the second is the number of subplot columns, and the third is the position where the subplot should be added.

```python
plt.figure()
plt.subplot(1, 2, 1)   #two rows of subplots, one column - create subplot 1
plt.plot(range(1,5),[7,3,9,1], color='red', marker='o', linestyle='dashed', linewidth=2, markersize=6)
plt.subplot(1, 2, 2)   #create subplot 2
plt.plot(range(1,5),[3,4,6,9], color='y', marker='*', linestyle='dashed', linewidth=0.5, markersize=12)
#plt.show()
```

The default axis ranges can be modified using the `xlim()` and `ylim()` functions.

```python
plt.figure()

plt.subplot(1, 2, 1)   
plt.ylim(0, 10)
plt.plot(range(1,5),[7,3,9,1], color='red', marker='o', linestyle='dashed', linewidth=2, markersize=6)

plt.subplot(1, 2, 2)
plt.ylim(0, 10)
plt.plot(range(1,5),[3,4,6,9], color='y', marker='*', linestyle='dashed', linewidth=0.5, markersize=12)

#plt.show()
```

We can make the same figure with slightly fewer lines of code, using the `subplots()` function.

```python
plt.figure()
fig, axs = plt.subplots(1, 2)   

axs[0].plot(range(1,5),[7,3,9,1], color='red', marker='o', linestyle='dashed', linewidth=2, markersize=6)
axs[0].set_ylim(0, 10)   #NB: we use set_ylim when calling on an `Axes` object

axs[1].plot(range(1,5),[3,4,6,9], color='y', marker='*', linestyle='dashed', linewidth=0.5, markersize=12)
axs[1].set_ylim(0, 10)

#plt.show()
```

With `subplots()`, we also have a bit more control over the plot details.

```python
from matplotlib.ticker import MultipleLocator

plt.figure()
fig, axs = plt.subplots(1, 2)   

axs[0].plot(range(1,5),[7,3,9,1], color='red', marker='o', linestyle='dashed', linewidth=2, markersize=6)
axs[0].set_ylim(0, 10)   #NB: we use set_ylim when calling on an `Axes` object

axs[1].plot(range(1,5),[3,4,6,9], color='y', marker='*', linestyle='dashed', linewidth=0.5, markersize=12)
axs[1].set_ylim(0, 10)

#Adjust the x and y ticks, for each plot
axs[0].xaxis.set_major_locator(MultipleLocator(0.5))
axs[0].yaxis.set_major_locator(MultipleLocator(1))
axs[1].xaxis.set_major_locator(MultipleLocator(0.5))
axs[1].yaxis.set_major_locator(MultipleLocator(1))

#plt.show()
```

Here is a complete example using `subplots()`:

```python
# Define some functions of x
x = np.linspace(-1,4,101)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.cos(2 * np.pi * x) * np.exp(-x)

# Create Figure and Axes objects, and set figure size
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the data
ax.plot(x,y1,"g-"  , label='sin($x$)') 
ax.plot(x,y2,"b--" , label='cos($x$)')
ax.plot(x,y3,"y-." , label='cos($2\pi x$) $e^{-x}$') # Use can use LaTeX here 

# Draw a legend
ax.legend(loc='best', fontsize='x-large',  prop={"size":20})

# Title the plot
ax.set_title('Example trigonometry functions', fontsize=20)

# Label the axes and increase the font
ax.set_xlabel("$x$"      , fontsize=20)
ax.set_ylabel("$y$"      , fontsize=20)

# Set the range of the x and y axis
ax.set_xlim(-1,4)
ax.set_ylim(-4.1,4.1)

# Set major axis multiple 
ax.xaxis.set_major_locator(MultipleLocator(0.5))
ax.yaxis.set_major_locator(MultipleLocator(1))

# Set minor axis multiple 
ax.xaxis.set_minor_locator(MultipleLocator(0.1))
ax.yaxis.set_minor_locator(MultipleLocator(0.1))

# Increase axis tick labels and flip them 
ax.tick_params(axis='both', which='major', length=8, labelsize=15,  direction = "in")
ax.tick_params(axis='both', which='minor', length=4, direction = "in")

# Draw a grid
ax.grid(True,which='major', axis='both')

# Save the plot as a png
fig.savefig("trig.png")
```

### A realistic example

This repository contains a dataset (`Data/stars.txt`) of the temperature and magnitude of various stars.  Let's make a scatter plot of this data.

First, we should load the data using `loadtext()`.

```python
data = np.loadtxt("https://raw.githubusercontent.com/jstupak/ComputationalPhysics/master/Data/stars.txt",float)

data
```

## Histograms

A histogram displays the "binned" frequency of observations.  The range of observable values is divided into discrete bins, each of which indicates the number of values that fall in the corresponding range.   Histograms are often used to estimate the probability distribution of a continuous variable, as well as counting experiments.

More information on Matplotlib histograms is available here: https://matplotlib.org/devdocs/api/_as_gen/matplotlib.pyplot.hist.html

To create a histogram, we simply have to pass a collection of values to the `hist()` function.

```python
values = [1.5, 2.5, 5.4, 2.3, 4.2, 5.1, 2.4, 0.2, 6.2, 9.5, 2.1, 2.6, 9.3, 2.7, 8.4, 9.4]

plt.figure()
plt.hist(values)
#plt.show()
```

In this example, Matplotlib determined a suitable $x$-axis range and binning.  Often, this is best left to a human.  We can specify the number of bins and axis range via the `bins` and `range` arguments.

```python
plt.figure()
plt.hist(values, bins=10, range=(0,10)) 
#plt.show()
```

It is sometimes desirable to have non-uniform bins.  To achieve this, pass a list of bin edges using the `bins` argument.

```python
plt.figure()
plt.hist(values, bins=[0, 0.5, 1, 2, 4, 6, 10])   #the range is determined based on the first/last bin edge
#plt.show()
```

The `hist()` function returns `array`s holding 1) the content of each bin, 2) the bin edges, and 3) the "patches" used to draw the histogram (generally, we don't care about the patches).

```python
plt.figure()

entries, bins, patches = plt.hist(values, bins=[0, 0.5, 1, 2, 4, 6, 10], density=True)
print("entries:",entries)
print("bins:",bins)
```

Several aspects of the histogram aesthetics can be adjusted using functions discussed above.

```python
plt.figure()
plt.hist(values, bins=[0, 0.5, 1, 2, 4, 6, 10])

plt.tick_params(axis='both', which='major', labelsize=12)
plt.xticks(range(10))
plt.xlabel("Observable"      , fontsize=12)
plt.ylabel("Entries"         , fontsize=12)

#plt.show()
```

To draw transparent histrograms, use the `histtype` argument.  The `lw` and `color` arguments can also be used to set the line width and color, respectively.

```python
plt.figure()
plt.hist(values, bins=[0, 0.5, 1, 2, 4, 6, 10], histtype='step', lw=2, color='red')

#plt.show()
```

This is particularly useful when we want to overlay two histograms.

```python
plt.figure()

fig, axs = plt.subplots(1, 2) 
axs[0].hist([2,2,5,5,5,8,8], bins=[0, 0.5, 1, 2, 4, 6, 10], lw=2, color='red')
axs[0].hist(values,          bins=[0, 0.5, 1, 2, 4, 6, 10], lw=2, color='green')

axs[1].hist([2,2,5,5,5,8,8], bins=[0, 0.5, 1, 2, 4, 6, 10], lw=2, color='red',   histtype='step')
axs[1].hist(values,          bins=[0, 0.5, 1, 2, 4, 6, 10], lw=2, color='green', histtype='step')

#plt.show()
```

Alternatively, use the `alpha` option to make the fill colors semi-transparent.

```python
plt.figure()

plt.hist([2,2,5,5,5,8,8], bins=[0, 0.5, 1, 2, 4, 6, 10], lw=2, color='red',   alpha = 0.5)
plt.hist(values,          bins=[0, 0.5, 1, 2, 4, 6, 10], lw=2, color='green', alpha = 0.5)
```

To produce a histogram representing the probability density for the observable, use the `density` option.

```python
plt.figure()

plt.hist(values, bins=[0, 0.5, 1, 2, 4, 6, 10], density=True)

#plt.show()
```

The `cumulative` option can be used to plot the number of entries in a given bin plus all smaller bins.

```python
plt.figure()

fig, axs = plt.subplots(1, 2) 
axs[0].hist(values, bins=[0, 0.5, 1, 2, 4, 6, 10])
axs[1].hist(values, bins=[0, 0.5, 1, 2, 4, 6, 10], cumulative=True)

#plt.show()
```

Choosing an appropriate axis range and number of bins is a bit subjective.  Too many/few bins makes it difficult to interpret a histogram.

```python
plt.figure()

fig, axs = plt.subplots(1, 2)
axs[0].hist(values, bins=100)
axs[1].hist(values, bins=2)

#plt.show()
```

#### Histograms with error bars

Drawing error bars on a histogram requires a bit of effort.  Suppose we want to add error bars to this histogram:

```python
plt.figure()
plt.hist(values, bins=[0, 0.5, 1, 2, 4, 6, 10])
```

We can accomplish this by drawing the histogram first, and then drawing error bars on top of the histogram.  To draw the error bars, we need a collection of the (central) $x$- and $y$-values, as well as the associated error.  The NumPy `histogram()` function (not to be confused with the Matplotlib `hist()` function) is useful for this.  This function doesn't actually produce a histogram.  It simply bins the data so it **could** be plotted, and returns arrays containing the bin content and edges.

```python
content,edges = np.histogram(values, bins=[0, 0.5, 1, 2, 4, 6, 10])

print("content:",content)
print("edges:",edges)
```

We draw error bars at the top of the histogram, so the $y$-values should correspond to the bin content, which we obtained above.  We draw error bars at the center of the bin, so the $x$-values should correspond to the bin centers.  We can get the bin centers from `edges` returned above. 

```python
centers = []
for i in range(len(edges[:-1])):
    centers.append((edges[i]+edges[i+1])/2)
    
centers
```

Assuming we are in the Gaussian regime, the uncertainty on the bin content is given by the square root of the bin content.  So we have all the ingredients now.

```python
plt.figure()
plt.hist(values, bins=[0, 0.5, 1, 2, 4, 6, 10], color="blue", edgecolor="red")
plt.errorbar(centers, content, yerr=np.sqrt(content),             color="red", linestyle='')

#plt.show()
```

## 2D Histograms

A 1D histogram represents the number of entries on a 2nd axis.  Similarly, a 2D histogram requires a 3rd axis to represent the number of entries.  Visualizing data in 3D is somewhat tricky.  

One remedy is to use a "color axis" to represent the 3rd dimension (a so-called "heat map" or "density plot").  This can be achieved using the `hist2d()` and `colorbar()` functions.

```python
plt.figure()

x = [1.5, 2.5, 5.4, 2.3, 4.2, 5.1, 2.4, 0.2, 6.2, 9.5, 2.1, 2.5, 9.3, 2.5, 8.4, 9.9]
y = [100, 123, 593, 203, 203, 506, 302, 605, 302, 593, 930, 502, 520, 203, 193, 390]

plt.hist2d(x, y, bins=(10,10))
plt.colorbar()
plt.show()
```

Let's revisit the scatter plot we made above using star data, this time using a 2D histogram to overcome graphical issues.

```python
data = np.loadtxt("../Data/stars.txt",float)
temp = data[:,0]   #all rows, column 0
mag  = data[:,1]

plt.hist2d(temp, mag, bins=(50,50))
plt.colorbar()
plt.xlabel("Temperature")
plt.ylabel("Magnitude")
plt.show()
```

The `cmap` argument can be used to specify an alternative color palette.

```python
plt.hist2d(temp, mag, bins=(50,50), cmap="magma")
plt.colorbar()
plt.xlabel("Temperature")
plt.ylabel("Magnitude")
plt.show()
```

The "Greys" palette is particularly useful for printing in greyscale.

```python
plt.hist2d(temp, mag, bins=(50,50), cmap="Greys")
plt.colorbar()
plt.xlabel("Temperature")
plt.ylabel("Magnitude")
plt.show()
```

We can even make the color axis logarithmic using the `norm` argument.

```python
from matplotlib import colors

plt.hist2d(temp, mag, bins=(50,50), cmap="Greys", norm=colors.LogNorm())
plt.colorbar()
plt.xlabel("Temperature")
plt.ylabel("Magnitude")
plt.show()
```

```python
from math import sqrt, sin, pi

wavelength = 5.0
k = 2*pi/wavelength
xi0 = 1.0
separation = 50.0      # Separation of centers in cm
side = 100.0           # Side of the square in cm
points = 500           # Number of grid points along each side
spacing = side/points  # Spacing of points in cm

# Calculate the positions of the centers of the circles
x1 = side/2 + separation/2
y1 = side/2
x2 = side/2 - separation/2
y2 = side/2

# Make an array to store the heights
xi = np.empty([points,points],float)

# Calculate the values in the array
for i in range(points):
    y = spacing*i
    for j in range(points):
        x = spacing*j
        r1 = sqrt((x-x1)**2+(y-y1)**2)
        r2 = sqrt((x-x2)**2+(y-y2)**2)
        xi[i,j] = xi0*sin(k*r1) + xi0*sin(k*r2)

# Make the plot
plt.figure()
plt.imshow(xi,origin="lower",extent=[0,side,0,side])
plt.gray()
plt.show()
```

## 3D Plots

Truly 3D plots can also be produced using Matplotilb.  To produce 3D `Axes`, simply call the `axes()` function with the argument string `projection='3d'`.

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

Just as in 2D, there are a variety of ways to plot in 3D.  

### Line graph

To produce a line graph, use the `plt3D()` function.

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

#generate data
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
 
#plot it
ax.plot3D(x, y, z, 'green')
#plt.show()
```

The plot formatting can be adjusted similarly to 2D plots.

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

#generate data
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
 
#plot it
ax.plot3D(x, y, z, 'green')
#plt.show()

ax.set_title('A 3D Line Plot', fontsize=20)

# Label the axes and increase the font
ax.set_xlabel("$x$", fontsize=15)
ax.set_ylabel("$y$", fontsize=15)
ax.set_zlabel("$z$", fontsize=15)

# Set the range of the x and y axis
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.set_zlim( 0,1)

# Set major axis multiple 
ax.xaxis.set_major_locator(MultipleLocator(0.5))
ax.yaxis.set_major_locator(MultipleLocator(0.5))
ax.zaxis.set_major_locator(MultipleLocator(0.25))

# Set minor axis multiple 
ax.xaxis.set_minor_locator(MultipleLocator(0.1))
ax.yaxis.set_minor_locator(MultipleLocator(0.1))
ax.zaxis.set_minor_locator(MultipleLocator(0.125))

# Increase axis tick labels and flip them 
ax.tick_params(axis='both', which='major', length=0, labelsize=10,  direction = "in")   #for 3D plots, setting axis='both' will affect all 3 dimensions
ax.tick_params(axis='both', which='minor', length=4, labelsize=10,  direction = "in")

# Draw a grid
ax.grid(True, which='major', axis='both')
```

### Surface plots

Surface plots require gridded data.  The easiest way to generate a grid of $x,y$ values is using the NumPy `meshgrid()` function.

```python
x = np.arange(-5, 5.25, 0.25)
y = np.arange(-2, 3, 1)

print("Before:")
print("x.shape =",x.shape)
print("y.shape =",y.shape)
print("\nx =",x)
print("y =",y)

x,y = np.meshgrid(x, y)

print("\nAfter:")
print("x.shape =",x.shape)
print("y.shape =",y.shape)
print("\nx =\n",x)
print("y =\n",y)
```

Then calculate the grid of $z$ values.

```python
r = np.sqrt(x**2 + y**2)
z = np.sin(r)

z
```

And finally plot the data.

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

surf = ax.plot_surface(x, y, z, cmap='viridis', antialiased=False)   #antialiasing improves the formatting slightly
fig.colorbar(surf)

#plt.show()
```

The grid above was intentionally sparse (for demonstration purposes).  A finer grid produces a smoother plot.

```python
x = np.arange(-5, 5.25, 0.25)
y = np.arange(-5, 5.25, 0.25)
x,y = np.meshgrid(x, y)

r = np.sqrt(x**2 + y**2)
z = np.sin(r)

fig = plt.figure()
ax = plt.axes(projection = '3d')

surf = ax.plot_surface(x, y, z, cmap='viridis', antialiased=False)   #antialiasing improves the formatting slightly
fig.colorbar(surf)

#plt.show()
```

## Animated Plots

The `FuncAnimation()` function can be used to produce animated plots.  It takes a `Figure` as argument, as well as a function that iteratively modifies the data.

First, generate the data and plot.  In this example, we will plot only the first $n$ data points, and gradually increase the value of $n$ until the full function is plotted.  So we should start with a plot without any data.

```python
#generate the data
t = np.linspace(0,2*np.pi)
x = np.sin(t)

#generate the plot (without data)
fig, ax = plt.subplots()
ax.axis([0,2*np.pi,-1,1])
line, = ax.plot([],[])   #plot() function returns a list of lines representing the plotted data
```

Then define a function that modifies the data.  Here, we define a function that produces a subset of the data via slicing.

```python
def animate(i):
    line.set_data(t[:i], x[:i])
```

Finally, make the animation.

```python
import matplotlib.animation
from IPython.display import HTML

ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(t))
HTML(ani.to_jshtml())
```
