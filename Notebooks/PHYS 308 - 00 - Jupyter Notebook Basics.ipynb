## What is a Jupyter Notebook?

A Jupyter Notebook is a web application (it opens in your browser) which allows you to write, save, and share interactive documents which contain text, live code (you can actually run the code within the browser), plots, images, and equations. Jupyter supports three default code languages (Julia, Python, and R) which is where its name comes from. You may also see these notebooks refered to as ipython notebooks but this is a somewhat depricated term. We will focus on python in this course.

More info: https://jupyter-notebook-beginner-guide.readthedocs.io/en/latest/what_is_jupyter.html

<font color=blue>Make and display professional-looking plots and graphs</font> 

```python
# Example plotting script using python 
#   as well as the numpy and matplotlib external packages 
import numpy as np
import matplotlib.pyplot as plt

# Define some functions
x = np.linspace(-1,4,101)
y1 = np.exp(-x)
y2 = -np.exp(-x)
y3 = np.cos(2 * np.pi * x) * np.exp(-x)

plt.plot(x,y1,"g--" , label='Damping rate')
plt.plot(x,y2,"g--" , label='')
plt.plot(x,y3,"b-." , label='Oscillation amplitude') # Use LaTeX 

# Set the range of the x and y axis
plt.xlim(0,4)
plt.ylim(-1.1,1.1)

# Label the axes and increase the font
plt.xlabel("time (s)"      , fontsize=20)
plt.ylabel("Amplitude (cm)"      , fontsize=20)

# Title the plot
plt.title('Under-damped Oscillation', fontsize=20)

# Draw a legend
#   loc -> location. examples: 'upper right', 'upper center', 'best'
#   fontsize. examples large etc.
#   framealpha -> set transparency
plt.legend(loc='upper right', fontsize='large', framealpha=1)

# Draw a grid
plt.grid(True)

# Increase axis tick labels (there are a few ways to do this)
plt.tick_params(axis='both', which='major', labelsize=20)

# Display the plot on screen 
plt.show()
```

<font color=blue>**Navigate your new notebook and create and run cells**</font> 

The Jupyter Notebook is organized into individual "cells". Initiall you have just one cell but you can create more by clicking on the "+" button on the toolbar.

There are two principal types of cells:
1. Code cells 
2. Markdown cells

You can switch between these using the dropdown menu on the toolbar. 

Code cells expect python code and they will interpret and run what you type after your press "shift enter"

Markdown cells are where you can write nicely formatted text (like in this current cell) and display it by pressing "shift enter". To learn how to edit text using markdown language you can click on "Help" then "Markdown".

Jupyter notebooks also allow you to navigate and create new cells using your keyboard and not your mouse.

When a cell is highlighted in <font color=green>green</font>  you are in "Edit Mode" and you can type and write code within the cell. To enter Edit Mode you just have to click on a cell with your mouse or move to it using your arrow keys and press "Enter"

When a cell is highlighted in <font color=blue>blue</font> you are in "Command Mode" and the keyboard can now be used to run shortcuts to create new cells, change the mode of a cell, delete cells, or move between cells. To enter Command Mode you can press "esc" on a currently highlighted green cell or by clicking outside of a cell with your mouse.

You can click on the toolbar "Help" then "Keyboard Shortcuts" to find out what commands are available when in Command Mode. My favorites are the following:

 - m - Convert a cell to a markdown cell
 - y - Convert a cell to a code cell
 - a - Insert new cell abbove
 - b - Insert new cell below
 - x - cut cell
 - c - copy cell
 - v - paste cell
 - Up and down arrow - navigate between cells
 - esc - Change to command mode
 - Enter - Change to edit mode
 - Shift-Enter - Run the cell
